## Approach
__Step 1__ Due to the fact that the dataset was originally stored in a TXT document, we chose to first read it using Pandas and then resave it as a CSV file for later use. If we choose N-fold cross validation, we will not consider resaving the partitioned dataset. Otherwise, we will choose to save the partitioned dataset for later use.
__Step 2__ After preparing the dataset, we need to perform some preprocessing on the dataset, such as removing null values (which actually exist in the dataset, and some reviews have empty text lines).
__Step 3__ After these preparations are completed, we need to perform undersampling on the dataset because it is extremely imbalanced, with the number of false reviews far less than the number of real reviews. If the problem of data imbalance is not addressed, the performance of most machine learning models is not very poor.This is because they often only learn the characteristics of real reviews without learning the characteristics of false reviews, and the ability to detect false reviews is what we want. In fact, there are many methods to solve data balancing. Considering the large amount of data in the model, we choose under-sampling.
__Step 4__ Then we need to perform feature engineering, and the reason we do so is that there are too few features provided in the dataset, and only three numerical types of features can be directly used. In fact, we can extract many useful features from other data provided by the dataset, which may require some domain expertise. Fortunately, many researchers have already proposed some useful features, and we can use them directly.After these tasks are completed, our data preparation work is considered basically completed.
__Step 5__ Afterwards, we will analyze and explore these features. During this process, we will draw bar charts to study the distribution of features in the overall dataset and use tree algorithms to analyze the importance of features. Afterwards, we will use different methods for feature selection and analyze the impact of different selection methods on the model's performance.
__Step 6__ Then there is the part of using different models to test the effectiveness. We are planning to use four models (__logistic regression__, __KNN__, __support vector machine__, and __naive Bayes__) for modeling separately. In the analysis of each model, we will conduct some hyperparameter debugging to find the model with better results. Afterwards, we will use these four models to build an ensemble model, and we will also compare the impact of different integration strategies (__voting method__ or __weighted average method__) on the model. Finally, we will compare our constructed ensemble model with the __random forest__ model (as it is a common ensemble learning model).Finally, if we have any additional energy, we will explore using pre-trained models such as BERT for false review recognition.At the end of the entire experiment, we will make a summary.
